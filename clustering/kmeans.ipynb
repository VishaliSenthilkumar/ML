```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
km = KMeans(n_clusters=3)
pred = km.fit_predict(X)
pred
```

```python
df0 = X[pred==0]
df1 = X[pred==1]
df2 = X[pred==2]

plt.scatter(df0[:,0],df0[:,1],c = 'yellow', label='setosa')
plt.scatter(df1[:,0],df1[:,1],c = 'blue', label='versicolor')
plt.scatter(df2[:,0],df2[:,1],c = 'green', label='virginica')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='red', marker='^', s=90, label='centroid')

plt.xlabel('setal length')
plt.ylabel('setal width')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng,sse)
```
